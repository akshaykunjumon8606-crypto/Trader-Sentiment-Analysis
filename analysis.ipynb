```python
sent=sent.drop_duplicates()
```

```python
trad=trad.drop_duplicates()
```

```python
sent['date']=pd.to_datetime(sent['date'])
sent['date']= sent['date'].dt.date
```

```python
trad.columns=trad.columns.str.lower().str.replace(" ","_")
```

```python
trad['timestamp_ist']=pd.to_datetime(trad['timestamp_ist'],dayfirst=True)
trad['date']=trad['timestamp_ist'].dt.date
```

```python
data=pd.merge(trad,sent,on="date",how="left")
data.head()
```

```python
pnl=data.groupby(['account','date'])['closed_pnl'].sum().reset_index()
```

```python
data['win']=data['closed_pnl'] > 0
rate=data.groupby('account')['win'].mean().reset_index()
```

```python
trades_day=data.groupby('date').size().reset_index(name='trade_count')
```

```python
data['side'].value_counts().plot(kind='bar',color='g')
plt.title("Long vs Short Trades")
plt.savefig("long_short_trades.png")
plt.show()
```

```python
sns.boxplot(x='classification', y='closed_pnl', data=data)
plt.title("Trader PNL VS Market Sentiment")
plt.xticks(rotation=45)
plt.savefig("pnl_vs_sentiment.png")
plt.show()
```

```python
sentiment_trade=data.groupby('classification').size().reset_index(name='trades')
```

```python
sns.barplot(x='classification',y='trades', data=sentiment_trade,color='c')
plt.title("Trade Frequency by Sentiment")
plt.xticks(rotation=45)
plt.savefig("trade_frequency_sentiment.png")
plt.show()
```

```python
sns.boxplot(x='classification',y='size_usd',data=data)
plt.title("Trade Size vs Sentiment")
plt.xticks(rotation=45)
plt.savefig("trade_size_sentiment.png")
plt.show()
```

```python
trade_count=data.groupby('account').size().to_dict()
```

```python
data['trader_type']=data['account'].map(lambda x:'Frequent' if trade_count[x] > 50 else 'Infrequent')
```

```python
pnl_trader=data.groupby('account')['closed_pnl'].sum()
```

```python
winner = pnl_trader[pnl_trader > 0]
print("Profitable Traders:",winner.count())
```
